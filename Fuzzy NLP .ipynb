```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Define the questions and answers
questions = [
    "What is TF-IDF vectorization?",
    "What is the purpose of TF-IDF?",
    "How does TF-IDF work?"
]

answers = [
    "TF-IDF (Term Frequency-Inverse Document Frequency) is a numerical representation of the importance of a word in a document. It is calculated as the product of term frequency and inverse document frequency.",
    "The purpose of TF-IDF is to assign a weight to each word in a document to reflect its importance in the document and in the corpus (collection of documents) as a whole.",
    "TF-IDF works by first calculating the term frequency of each word in a document. This represents how frequently the word appears in the document. The inverse document frequency is then calculated as the logarithm of the total number of documents divided by the number of documents containing the word. The final TF-IDF weight is obtained by multiplying the term frequency and inverse document frequency."
]

# Create a TfidfVectorizer object
vectorizer = TfidfVectorizer()
# Fit the vectorizer to the questions
X = vectorizer.fit_transform(questions)
# print(X)
# Define a function to get the most similar question to a given input
def get_answer(input_question):
    input_vector = vectorizer.transform([input_question])
#     print(input_vector)
    cosine_similarities = np.dot(input_vector, X.T).toarray()[0]
    print(cosine_similarities)
    most_similar_index = cosine_similarities.argmax()
    print(most_similar_index)
    return answers[most_similar_index]



# Test the chatbot
input_question = "What does TF-IDF measure?"
print(get_answer(input_question))
```
